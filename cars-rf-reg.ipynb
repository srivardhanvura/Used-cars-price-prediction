```python
def onehot_encode(car,columns,prefixes):
  car=car.copy()
  for column,prefix in zip(columns,prefixes):
    dummies = pd.get_dummies(car[column],prefix=prefix)
    car=pd.concat([car,dummies],axis=1)
    car=car.drop(column,axis=1)
  return car
```

```python
company=[]
for x in car['Name']:
    company.append(x.split(' ')[0])

car['Company']=company
car.drop('Name',axis=1,inplace=True)
car
```

```python
y=car.loc[:,'Price']
x=car.drop('Price',axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x=scaler.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 200,random_state=1)
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
