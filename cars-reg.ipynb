<h3>Reading data<h3>

<b>Determining percentage of null values in each column<b>

```python
print(data.shape)
print(data.isnull().sum()/data.shape[0]*100)
```

Dropping columns with null value percentage greater than 70

```python
cols = [col for col in data.columns if (data[col].isnull().sum()/data.shape[0] * 100 < 70)]
data_trimmed = data[cols]

print ('Columns after dropping null columns: %s' % data_trimmed.shape[1])
```

```python
 data_trimmed.isnull().sum()/data.shape[0]*100
```

```python
data_trimmed.head()
```

Converting features to float dtype

```python
data_trimmed.dtypes
```

```python
data_trimmed.fillna(data.mode().iloc[0])
data_trimmed.isnull().sum()
```

```python
data_trimmed['Mileage'] = pd.to_numeric(data_trimmed['Mileage'].str.lower().str.split().str.get(0), errors='coerce')
data_trimmed['Engine'] = pd.to_numeric(data_trimmed['Engine'].str.lower().str.split().str.get(0), errors='coerce')
data_trimmed['Power'] = pd.to_numeric(data_trimmed['Power'].str.lower().str.split().str.get(0), errors='coerce')
```

```python
data_trimmed.dtypes
```

```python
data_trimmed.head()
```

```python
cars=data_trimmed['Name'].str.lower().unique()
```

```python
num_cars=len(cars)
num_cars
```

Adding a new column with the name Company

```python
comp = data_trimmed['Name'].str.split().str.get(0)

data_trimmed.insert(0, "Company", comp, True)
data_trimmed
```

```python
name_dummies = pd.get_dummies(data_trimmed.Name, drop_first=True)
company_dummies = pd.get_dummies(data_trimmed.Company, drop_first=True)
location_dummies = pd.get_dummies(data_trimmed.Location, drop_first=True)
fuel_dummies = pd.get_dummies(data_trimmed.Fuel_Type, drop_first=True)
transmission_dummies = pd.get_dummies(data_trimmed.Transmission, drop_first=True)
owner_dummies = pd.get_dummies(data_trimmed.Owner_Type, drop_first=True)
seats_dummies = pd.get_dummies(data_trimmed.Seats, drop_first=True)
```

```python
features = data_trimmed.drop(['Name', 'Company', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Seats'], axis = 1)
```

```python
dfle = pd.concat([name_dummies, location_dummies, features, fuel_dummies, transmission_dummies, owner_dummies, seats_dummies], axis = 1)
dfle
```

Applying Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

dfle = dfle.dropna()
X = dfle.drop('Price', axis = 1)
X.Year=2020-X["Year"]
y = dfle.Price
X_copy=X

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
print(model.score(x_test,y_test))
y_pred=model.predict(x_test)
```
