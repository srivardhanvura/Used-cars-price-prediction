```python
def onehot_encode(car,columns,prefixes):
  car=car.copy()
  for column,prefix in zip(columns,prefixes):
    dummies = pd.get_dummies(car[column],prefix=prefix)
    car=pd.concat([car,dummies],axis=1)
    car=car.drop(column,axis=1)
  return car
```

```python
car=car.drop('Name',axis=1)
```

```python
car=onehot_encode(
     car,['Location','Fuel_Type','Transmission','Owner_Type'],
           ['Loc','Fuel','Trans','OwnerT']
)
```

```python
car['Engine']=car['Engine'].str.split(' ').str.get(0).str.replace(',','')
car['Power']=car['Power'].str.split(' ').str.get(0).str.replace(',','')
car['New_Price']=car['New_Price'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car=onehot_encode(
     car,['Engine','Power','New_Price'],
           ['engine','Pow','New_p']
)
```

```python
y=car.loc[:,'Price']
x=car.drop('Price',axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x=scaler.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
car.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 200)
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)*100
```
