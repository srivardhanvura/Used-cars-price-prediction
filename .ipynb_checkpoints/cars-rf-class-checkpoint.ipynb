```python
df = df.fillna(df.mode().iloc[0])
df.head()
```

```python
df['Price'] = df['Price'].round(0).astype(int)
inputs = df.drop('Price',axis='columns')
```

```python
target=df['Price']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Name = LabelEncoder()
le_Location = LabelEncoder()
le_Year = LabelEncoder()
le_Kilometers_Driven = LabelEncoder()
le_Fuel_Type = LabelEncoder()
le_Transmission = LabelEncoder()
le_Owner_Type = LabelEncoder()
le_Mileage= LabelEncoder()
le_Engine = LabelEncoder()
le_Power= LabelEncoder()
le_Seats = LabelEncoder()
le_New_Price = LabelEncoder()
```

```python
inputs['Name_n'] = le_Name.fit_transform(inputs['Name'])
inputs['Location_n'] = le_Name.fit_transform(inputs['Location'])
inputs['Year_n'] = le_Name.fit_transform(inputs['Year'])
inputs['Kilometers_Driven_n'] = le_Name.fit_transform(inputs['Kilometers_Driven'])
inputs['Fuel_Type_n'] = le_Name.fit_transform(inputs['Fuel_Type'])
inputs['Transmission_n'] = le_Name.fit_transform(inputs['Transmission'])
inputs['Owner_Type_n'] = le_Name.fit_transform(inputs['Owner_Type'])
inputs['Mileage_n'] = le_Name.fit_transform(inputs['Mileage'])
inputs['Engine_n'] = le_Name.fit_transform(inputs['Engine'])
inputs['Power_n'] = le_Name.fit_transform(inputs['Power'])
inputs['Seats_n'] = le_Name.fit_transform(inputs['Seats'])
inputs['New_Price_n'] = le_Name.fit_transform(inputs['New_Price'])
inputs.head()
```

```python
inputs_n = inputs.drop(['Name','Location','Year','Kilometers_Driven','Fuel_Type','Transmission','Owner_Type','Mileage','Engine','Power','Seats','New_Price'],axis ='columns')
inputs_n
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix


x_train, x_test, y_train, y_test = train_test_split(inputs_n,target,test_size=0.2)
```

```python
len(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 200)
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
